```python
import matplotlib.pyplot as plt

y1 = [1, 3, 4, 56, 67]
y2 = [3, 4, 6, 7, 9]

plt.plot(y1)
plt.plot(y2)

plt.xlim(0, len(y1)-1) # set the x-axis range from 0 to the length of y1 minus 1

plt.show()
```

```python
import numpy as np

def sqar(x):
  
    return x**2

x = np.linspace(-5,5, 100)
y = sqar(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2')
plt.show()
```

## scatter plot 

```python
import matplotlib.pyplot as plt
X = [[1, 2, 3, 4, 1], [1, 2, 3, 4, 5]]
Y = [0, 1, 0, 1, 1]
plt.scatter(X[0], X[1], c=Y, s=40, cmap=plt.cm.Spectral)
plt.show()
```

```python
# c = color , s = size, 
# c = Y , means it will give 2 different color base on the different value of Y
```
