# 2. Probability Denstity funciton (PDF)

<img src='images/pdf.jpg' width=75%>


```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde

# Create a list of values
values = [13495.0, 16500.0, 16500.0, 13950.0, 17450.0, 15250.0]

# Estimate the PDF using a Gaussian kernel density estimator
pdf = gaussian_kde(values)

# Create a range of values to plot the PDF
x = np.linspace(min(values), max(values), 100)

# Plot the PDF
plt.plot(x, pdf(x))
plt.title("Probability Density Function")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
